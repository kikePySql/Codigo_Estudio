# Condicionales

1. IF
2. ELIF
3. ELSE

```python
# Determinar un algoritmo que identifique cual de los dos números es mayor, menor o igual
x = int(input("Ingrese el valor para la variable x: "))
y = int(input("Ingrese el valor para la variable y: "))

if x < y:
  print(f"El número {x} es menor que {y}")
elif x == y:
  print(f"El número {x} es igual que {y}")
else:
  print(f"El número {x} es mayor que {y}")
```

# Ejercicio para clasificar rango de edades

```python
edad = int(input("Ingrese la edad: "))

if edad < 12:
  print("Niño")
elif edad >= 12 and edad < 18:
  print("Adolecente")
elif edad >= 18 and edad < 60:
  print("Adulto")
elif edad >= 60 and edad < 120:
  print("Adulto Mayor")
else:
  print("Se presento un error, lo invitamos amablemente a retomar el ejercicio")
```

# Ciclo For

Sintaxis:

for iterador in secuencia:
  if condicional:
    continue
  codigo a ejecutar si la condición no se cumple

Ejemplo 1. Aplicación de ciclo for (para)

```python
for numero in range(1,11):
  print(numero)
```

Ejemplo 2. Aplicación de la declaración **continue**

Imprimir los números del 1 al 10, omitiendo el número 5

```python
for numero in range(1,11):
  if numero == 5:
    continue # Salta la iteración cuando el iterador número es igual a 5 y continua
  print(numero)
```

# Ciclo While (Mientras)




Ejemplo 3. Muestre los primeros

```python
contador = 1
while contador <= 10:
  print(contador)
  # contador = contador + 1
  contador += 1 # Se aumenta la variable contador
```

Ejemplo 4. Contar del 1 al 10, omitiendo el número 7

```python
contador = 1
while contador <= 10:
  if contador == 7:
    contador = contador + 1 # Se aumenta la variable contador y se contua con la
    #siguiente iteración, es decir que contador queda con el valor de 8 y se muestra
    continue
  print(contador)
  contador += 1 # Se aumenta la variable contador
```

Ejercicio práctico 2. Suponga que se está procesando una lista de estudiantes y se quiere omitir a aquellos que tienen menos de 60 puntos. Realice un código que le dé solución al planteamiento

```python
estudiantes = {"Felipe":65, "Maria":55, "Enrique": 90, "Andrea": 40}

print("Estudiantes aprobados:")
for estudiante, puntaje in estudiantes.items():
  if puntaje < 60:
    continue # se omiten los estudiantes que no aprobaron
  print(f"{estudiante} ha aprobado con {puntaje} puntos.")
```
