# Ejercicio 4.
4. Condicional: Pedir IMC (Indice de Masa Coproral) = Masa/(Estatura)**2
Datos de entrada: Masa y estatura
Clasificación del IMC: x menor 18.5 **Bajo peso**, 18.5 menor o igual 24.9 **Peso saludable**, 25.9 menor o igual 29.9 **Sobre peso**, mayor a 30 **Kilos moratales**

```python
masa = float(input("Por favor ingrese su masa corporal: "))
estatura = float(input("Por favor ingrese su estatura: "))

IMC = masa/((estatura)**2)

if IMC < 18.5:
  print("Bajo peso")
elif IMC >= 18.5 and IMC <=24.9:
  print("Peso saludable")
elif 25.0 <= IMC <= 29.9:
  print("Sobre peso")
else:
  print("Kilos mortales")
```

# Ejercicio 5

a. Hacer el punto anterior 100 veces (for)

b. Hacer el punto anterior cada que el usuario así lo desee (while)

```python
veces = int(input("Por favor ingrese el número de datos que quiere calcular: "))
bajo_peso = 0
bajo_peso_lista = []
peso_saludable = 0
peso_saludable_lista = []
sobre_peso = 0
sobre_peso_lista = []
kilos_mortales = 0
kilos_mortales_lista = []


for i in range(veces):
  nombre = str(input("Ingrese su nombre: "))
  masa = float(input("Por favor ingrese su masa corporal: "))
  estatura = float(input("Por favor ingrese su estatura: "))

  IMC = masa/((estatura)**2)

  if IMC < 18.5:
    print("Bajo peso")
    bajo_peso_lista.append(nombre)
    bajo_peso +=1  # =bajo_peso=bajo_peso+1. Es un contador y aquí se usa como acumulador
  elif IMC >= 18.5 and IMC <=24.9:
    print("Peso saludable")
    peso_saludable_lista.append(nombre)
    peso_saludable +=1
  elif 25.0 <= IMC <= 29.9:
    print("Sobre peso")
    sobre_peso_lista.append(nombre)
    sobre_peso +=1
  else:
    print("Kilos mortales")
    kilos_mortales +=1
    kilos_mortales_lista.append(nombre)

print(f"Nombre de personas con bajo peso: {bajo_peso_lista}")
print(f"Bajo peso: {bajo_peso}")
print(f"Nombre de personas con peso saludable: {peso_saludable_lista}")
print(f"Peso saludable: {peso_saludable}")
print(f"Nombre de personas sobre peso: {sobre_peso_lista}")
print(f"Sobre peso: {sobre_peso}")
print(f"Nombre de personas con kilos mortales: {kilos_mortales_lista}")
print(f"Kilos mortales: {kilos_mortales}")
```
