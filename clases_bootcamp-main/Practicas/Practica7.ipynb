# Indexación y slicing en arrays NumPy:

Puedes acceder a elementos individuales de un array NumPy utilizando la indexación, y también puedes realizar #slicing" para seleccionar subconjuntos de elementos.

Ejemplo de indexación y slicing en un array NumPy

```python
# Indexación
import numpy as np

array3 = np.array([0,1,2,3,4,6,7])
print("Mi elemento en la posición 1:", array3[1])
```

```python
# slicing
sub_array = array3[2:8] # Rango de posiciones
sub_array
```

# Ejemplos de operaciones matriciales con NumPy:

NumPy es especialmente poderoso cuando se trata de operaciones matriciales. Puedes realizar multiplicación de matrices, transposición, cálculos de determinantes y mucho más de manera efiente.

Ejemplo de multiplicación de matrices con NumPy:

```python
matriz1 = np.array([[1,2],[3,4]])
matriz1
```

```python
matriz2 = np.array([[5,6],[7,8]])
matriz2
```

# Problema: Cálculo de Estadística en Datos de Ventas

Supongamos ue tienes un conjunto de datos de ventas de productos en una tienda durante varios meses. Cada fila del conjunto de datos representa una venta y contiene información sobre el producto vendido, la cantidad vendida y el precio unitario. Tu objetivo es realizar algunas estadísticas básicas con estos datos utilizando NumPy.

```python
# datos de venta (producto, cantidad, precio unitario)

ventas = np.array([
    ["Shampoo",       100,  24000],
    ["Acondicionador", 50,  13000],
    ["Jabón",          75,  6000 ],
    ["Crema de manos", 120, 18000],
    ["Protector solar", 60, 82000]
])

# Calcular el total de ventas para cada producto

total_ventas = ventas[:,1].astype(float)* ventas[:,2].astype(float)
print(total_ventas)

# Calcular el total de ventas en general
total_ventas_general = np.sum(total_ventas)
print(total_ventas_general)

# Calcular el producto más vendido
producto_mas_vendida = ventas[np.argmax(ventas[:,1].astype(int)),0]
print(producto_mas_vendida)

# Calcular el producto menos vendido
producto_menos_vendido = ventas[np.argmin(ventas[:,1].astype(int)),0]
print(producto_menos_vendido)

# Calcular valor promedio por producto
precio_promedio_por_productos = np.mean(ventas[:,2].astype(float))
print(precio_promedio_por_productos)

# Calcular el valor que corresponde a la madiana del precio del producto
mediana_precio_producto = np.median(ventas[:,2].astype(float))
print(mediana_precio_producto)
```
